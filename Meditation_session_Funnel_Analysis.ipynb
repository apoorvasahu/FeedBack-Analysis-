```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS 
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
# %matplotlib implot
```

```python
data=pd.read_excel("/home/apoorva/Downloads/Meditation raw data_0110.xlsx", sheet_name='Feedback')
```

```python
df = pd.DataFrame(data, columns= ['Answer'])
```

STEP 1 - Extracting words 

```python
string=df.Answer.tolist()
string
```

```python
a = ' '.join([str(elem) for elem in string]) 
a
```

```python
tokenized_words=word_tokenize(a, "english")
```

```python
allWords=' '.join([r for r in tokenized_words])
```

```python
plt.figure(figsize=(14,7))

wordCloud=WordCloud(width=500, height =300, random_state=21, max_font_size=119).generate(allWords)
plt.imshow(wordCloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

STEP 2 - Calculating Positive/Negative/Neutral Rating

```python
data
```

```python
data['Answer']=data['Answer'].astype(str)
```

```python
data.drop(['Unnamed: 9'],axis=1,inplace=True)
#removing empty columns
```

POSITIVE RATING

```python
positive=data[(data['Answer']=="5") | (data['Answer']=="4")].groupby('Session id')['Answer'].count().reset_index()
positive
```

Session ID 79 performing best

NEGATIVE RATING

```python
negative=data[(data['Answer']=="1") | (data['Answer']=="2")].groupby('Session id')['Answer'].count().reset_index()
negative.head()
```

Note - There is no where rating below than 3

NEUTRAL RATING

```python
neutral=data[data['Answer']=="3"].groupby('Session id')['Answer'].count().reset_index()
neutral.head()
```

Merging to calculate NPS


```python
rating_nps=pd.merge(positive,neutral,on='Session id', how='outer')
rating_nps.head()
```

```python
rating_nps.columns = ['Session id', 'Promoters', 'Passives']
```

```python
rating_nps['Passives'] = rating_nps['Passives'].fillna(0)
#replacing null values with zero for easy calculation for NPS
```

```python
rating_nps.head()
```

Calculating NPS = (Positive - Negative)/ Total responses

```python
rating_nps['NPS']=rating_nps['Promoters']/(rating_nps['Passives']+rating_nps['Promoters'])
```

```python
rating_nps.head()
```

Step 3- Writing DataFrame to excel file.



```python
rating_nps.to_excel("/home/apoorva/Downloads/Meditation.xlsx",
             sheet_name='Quantity_NPS') 
```

Step 4- Calculating Positive/Negative/Neutral Quaitative Feedback. Calculating NPS



```python
data2=pd.read_excel("/home/apoorva/Downloads/Meditation_raw.xlsx", sheet_name='FeedBack_Qualitative_Analysis')
```

```python
data2.drop(['Category'],axis=1,inplace=True)
```

```python
data2
```

```python
promoter=data2[data2['Positive']==True].groupby('Session id')['Positive'].value_counts()
promoter
```

```python
detractor=data2[data2['Negative']==True].groupby('Session id')['Negative'].value_counts()
detractor
```

```python
passives=data2[data2['Neutral']==True].groupby('Session id')['Neutral'].value_counts()
passives
```

Merging first 2 dataframes

```python
h=pd.merge(promoter,detractor,on='Session id', how='outer')
h
```

```python
quality=pd.merge(h,passives,on='Session id', how='outer')
quality
```

79 is the best session id since it has most positive rating qualitatively and quantitatively

```python
quality['Positive'] = quality['Positive'].fillna(0)
quality['Negative'] = quality['Negative'].fillna(0)
quality['Neutral'] = quality['Neutral'].fillna(0)
```

Calculating NPS

```python
quality['NPS']=(quality['Positive']-quality['Negative'])/(quality['Positive']+quality['Negative']+quality['Neutral'])
```

```python
quality
```

Step 5- Writing DataFrame to excel file.



```python
quality.to_excel("/home/apoorva/Downloads/Meditation.xlsx",
             sheet_name='Quality_NPS') 
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(quality.corr())
```
